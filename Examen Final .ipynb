#### ¿Qué entidad tiene mayor número muertes? 

```python
# se revisa la xolumna de Descripcion de la Entidad y cuantas veces se repite cada una, con la funcion value_count()
Entidades = MortalidadMaterna['Descripción de la entidad de ocurrencia']
CuantasMuertes = len(Entidades)
Entidad_mas_muerte = Entidades.value_counts()  # se obtiene el numero de la Entidad que mas muertes regstro
MasMuerteRegistrada = Entidad_mas_muerte.max()  # se obtiene el total numero de muertes registrado en la Entidad con mas muertes

print('La Entidad que mas muertes registro es la ',Entidad_mas_muerte)
print()
print('con un numero de ', MasMuerteRegistrada, ' de', CuantasMuertes,'registradas')
```
